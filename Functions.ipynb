```python
liste = [1,2,3,4,5,6,7,8,9,99]
def maximum(liste):
    for i in range(0, len(liste)-1):
        if liste[i]> liste[i+1]:
            maxi = liste[i]
        else:
            maxi = liste[i+1]
    return(maxi)
maximum(liste)
```

```python
listee = [52,70,97,20,75,62]

#sum function
#------------
def qsum(listee):
    s = 0
    for i in listee:
        s = s + i
    return(s)


qsum(listee)

#multiplication function
#-----------------------
def qmult(listee):
    m = 1
    for i in listee:
        m = m * i
    return(m)


qmult(listee)


#Final function
#--------------
oddlist = list()
evenlist = list()

for i in range(len(listee)):
    if i % 2 == 1:
        i = oddlist.append(listee[i])
    elif i % 2 == 0:
        i = evenlist.append(listee[i])

print("oddlist =",oddlist)
print("evenlist =",evenlist)

print("multiplication of odd values:",qmult(oddlist))
print("sum of even values:",qsum(evenlist) )
```

```python
import math

def quest4(D):
    H = 30
    C = 50
    templist = []
    for d in D:
        Q = round((math.sqrt((2*C*int(d))/H)))
        templist.append(Q)
    return(templist)

templist = input()
templist = templist.split(",")
quest4(templist)
```
