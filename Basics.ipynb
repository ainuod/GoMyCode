```python
number=int(input("Enter an integer:"))
fact=1
if (number < 0):
    print("Negative numbers do not have a factorial.")
elif (number == 0):
    fact=1
    print("The factorial of 0 is equal to 1 (by convention).")
else:
    for i in range(number,0,-1):
        fact=fact*i
    print("The factorial of ",number," is equal to ",fact,".")
```
